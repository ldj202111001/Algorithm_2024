# 5. 알고리즘 코드

```python
def counting_sort(A):
    output = [0] * len(A) # 정렬된 결과 리스트 임시 저장
    count = [0] * MAX_VAL # 각 킷값의 빈도를 저장하는 인덱스


    for i in A:
        count[i] += 1     # 리스트 data의 각 킷값의 빈도수 계산


    for i in range(MAX_VAL):
        count[i] += count[i-1] # 전 빈도수를 다음빈도수에 더한다 즉, 빈도테이블을 위치 테이블로 전환하는 과정이다.
                               # 위치 테이블: 인덱스 = 킷값, 값 = 결과 리스트[값] 이전까지 인덱스 저장
    for i in range(len(A)):
        output[count[A[i]] - 1] = A[i] # 위치 테이블을 토대로 결과 리스트에 리스트 정렬
        count[A[i]] -= 1


    for i in range(len(A)):
        A[i] = output[i]    # 결과 리스트를 원래 리스트에 저장

       


MAX_VAL = 10
data = [1, 4, 1, 2, 7, 5, 2]
print("Original : ", data)
counting_sort(data)
print("Counting : ", data)
```

# 8. 복잡도 분석
### 시간 복잡도:  
입력의 크기가 n일때, 숫자의 범위가 k가지 라면,  
입력 리스트의 빈도수 계산을 위해 모든 리스트를 한번씩 순회해야 하므로 n번,  
위치 테이블 구현을 위해 숫자의 범위 만큼 순회해야 하므로 k번,
전체 알고리즘은 O(k+n) 즉, O(n) 의 시간복잡도를 갖는다.
