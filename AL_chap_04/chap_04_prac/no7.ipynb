# 문제정의
다음의 방향 그래프에 대해 DFS 기반 위상 정렬을 수행하라.   


```python
graph1 = {
           'A' : {'B', 'D', 'E'},
           'B' : {'C', 'E', 'F'},
           'C' : {'G'},
           'D' : {'E'},
           'E' : {},
           'F' : {'G'},
           'G' : {}
           
}

graph2 = {
           'A' : {'B', 'C'},
           'B' : {'D'},
           'C' : {'D', 'E'},
           'D' : {'F'},
           'E' : {'G', 'H'},
           'F' : {},
           'G' : {'H'},
           'H' : {'C'}
}

def topological_sort(graph): # 매개변수: 그래프
    inDeg = {} # 진입차수(딕셔너리)
    for v in graph:
        inDeg[v] = 0       # 모든 진입차수 0으로 초기화
    for v in graph:        # 모든 그래프 반복문
        for u in graph[v]: # 그 그래프의 인접정점이 있을경우
            inDeg[u] += 1  # 해당 인접정점의 진입차수를 +1     

 # 여기까지는 그래프에대한 진입차수를 초기화

    vlist = [] # 진입차수가 0인 리스트
    for v in graph: # 모든 정점에 대해
        if inDeg[v] == 0: # 진입 차수가 0 이라면
            vlist.append(v) #진입차수가 0인 리스트에 추가

    while vlist: # 종료조건: vlist가 공백일때
        v = vlist.pop() # v: 진입차수 0인 정점
        print(v, end = '') # v 출력
        for u in graph[v]: # 정점 v의 인접정점들 중에서
            inDeg[u] -= 1  # 인접차수 -1
            if inDeg[u] == 0: # 만약 인접차수가 0 이라면
                vlist.append(u) # 다시 진입차수가 0인 리스트에 넣기

print('graph1: ')
topological_sort(graph1)
print('\ngraph2: ')
topological_sort(graph2)
```

graph2 는 C - E - G - H 가 사이클이다. 사이클의 모든 정점들은 선행 정점을 갖게되고 인접차수가 줄어들수 없는 구조가된다.   
따라서 사이클 C - E - G - H 와 그와 인접한  D, F 는 출력이 된지 않는다.

*진입차수: 한 정점이 갖고있는 선행 정점의 개수
           한 정점에서 다른 정점으로 가기 위해서는 진입차수 = 0 이 충족되야한다.
