```python
graph1 = {
           'A' : {'B', 'D', 'E'},
           'B' : {'C', 'E', 'F'},
           'C' : {'G'},
           'D' : {'E'},
           'E' : {},
           'F' : {'G'},
           'G' : {}
           
}

graph2 = {
           'A' : {'B', 'C'},
           'B' : {'D'},
           'C' : {'D', 'E'},
           'D' : {'F'},
           'E' : {'G', 'H'},
           'F' : {},
           'G' : {'H'},
           'H' : {'C'}
}

def topological_sort(graph): # : 
    inDeg = {} # ()
    for v in graph:
        inDeg[v] = 0       #   0 
    for v in graph:        #   
        for u in graph[v]: #    
            inDeg[u] += 1  #    +1     

 #    

    vlist = [] #  0 
    for v in graph: #   
        if inDeg[v] == 0: #   0 
            vlist.append(v) # 0  

    while vlist: # : vlist 
        v = vlist.pop() # v:  0 
        print(v, end = '') # v 
        for u in graph[v]: #  v  
            inDeg[u] -= 1  #  -1
            if inDeg[u] == 0: #   0 
                vlist.append(u) #   0  

print('graph1: ')
topological_sort(graph1)
print('\ngraph2: ')
topological_sort(graph2)
```
